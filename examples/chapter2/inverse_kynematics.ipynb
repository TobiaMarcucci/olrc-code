```python
import numpy as np
import cvxpy as cp
import matplotlib.pyplot as plt
```

```python
# Numerical parameters.
n = 3 # number of links
l = 1 # link lengths
p = np.array([2, 1]) # desired gripper position
```

```python
# Forward kinematics: position of gripper as a funciton of configuration vector.
def fk(q):
    x = cp.sum(l * cp.cos(cp.sum(q[:i+1])) for i in range(q.size))
    y = cp.sum(l * cp.sin(cp.sum(q[:i+1])) for i in range(q.size))
    return cp.hstack([x, y])
```

```python
# Variables.
q = cp.Variable(n)

# Constraints.
constraints = [fk(q) == p]

# Objective function: each angle must be close to pi/2.
obj = cp.sum_squares(cp.cos(q))

# Initial guess.
q.value = np.zeros(q.size)

# Solve problem.
prob = cp.Problem(cp.Minimize(obj), constraints)
prob.solve(nlp=True, solver=cp.IPOPT)
```

```python
# Plot solution.
plt.figure()
plt.gca().set_aspect('equal')
robot_points = np.vstack([fk(q.value[:i]).value for i in range(n + 1)])
plt.plot(*robot_points.T, marker='o')
plt.grid()
```
