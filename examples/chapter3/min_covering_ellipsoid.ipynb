```python
import numpy as np
import cvxpy as cp
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
```

```python
# Points to be contained by the ellipsoid
np.random.seed(0)
n = 2
m = 100
x = np.random.rand(m, n)
```

```python
# Ellipsoid matrix and offset
A = cp.Variable((n, n), PSD=True)
b = cp.Variable(n)

# Semidefinite program
objective = - cp.log_det(A)
constraints = [cp.norm2(A @ xi + b) <= 1 for xi in x]
prob = cp.Problem(cp.Minimize(objective), constraints)
prob.solve()
```

```python
if n == 2:

    # Get ellipse center and angle
    c = - np.linalg.solve(A.value, b.value)
    P = A.value.T @ A.value
    eigvals, eigvecs = np.linalg.eigh(P)
    axes = 2 / np.sqrt(eigvals)
    angle = np.degrees(np.arctan2(eigvecs[1, 0], eigvecs[0, 0]))
    
    # Plot ellipse.
    plt.figure()
    plt.scatter(*x.T, marker='x')
    plt.gca().add_patch(Ellipse(c, axes[0], axes[1], angle=angle, ec='k', fc='None'))
    plt.show()
```
