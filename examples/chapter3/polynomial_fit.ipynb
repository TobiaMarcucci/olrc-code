```python
import numpy as np
import cvxpy as cp
import matplotlib.pyplot as plt
```

```python
# Data points
np.random.seed(0)
n = 100
noise = .3
x = np.linspace(0, 4 * np.pi, n)
y_nom = np.sin(x)
y = y_nom + np.random.uniform(-noise, noise, n)
```

```python
# Degree of polynomial
d = 7
eps = 2 * noise

# Monomials evaluated at each data point (Vandermonde matrix)
X = np.vstack([x ** i for i in range(d + 1)]).T

# Optimization variable (coefficients)
c = cp.Variable(d + 1)

# Quadratic program
resdiuals = X @ c - y
objective = cp.sum_squares(resdiuals)
constraints = [cp.abs(resdiuals) <= eps]
prob = cp.Problem(cp.Minimize(objective), constraints)
prob.solve()

# Print status
print('Problem status:', prob.status)
print('Problem optimal value:', prob.value)
```

```python
# Plot fitting function
plt.figure()
plt.scatter(x, y, label='Data points')
plt.plot(x, X @ c.value, c='k', label='Polynomial fit')
plt.plot(x, X @ c.value + eps, c='red', label='Max error')
plt.plot(x, X @ c.value - eps, c='red')
plt.legend()
plt.show()
```
